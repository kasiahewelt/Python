## Praca domowa nr 13

Podziel zbiór na treningowy i testowy w proporcji 2:1 [10%]

```python
X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size = 0.6 )
```

Stwórz nowy Forest Regressor z domyślnymi parametrami i dopasuj go do zbioru treningowego [10%]

```python
rf_regressor = RandomForestRegressor()
rf_regressor.fit(X_train, y_train)
```

Narysuj jedno drzewo tworzące powyższy las losowy, użyj atrybutu .estimators_ oraz funkcji plot_tree [10%]

```python
rf_regressor.estimators_

plt.figure(figsize = (15,12))
plot_tree(rf_regressor.estimators_[0], feature_names=boston.feature_names)#, class_names=boston.target_names)

```

Sprawdz score na danych testowych [10%]

```python
rf_regressor.score(X_test, y_test)
```

* Sprawdz jak zmienia się score gdy liczba estymatorów zmienia się od 1 do 100 (parametr n_estimators) [20%]

Narysuj wykres zależności wyniku od wartości parametru. Jakie masz wnioski?

```python
estimators = list(range(1,101))
scores =[]
for n_est in estimators:
    #print(n_est)
    rf= RandomForestRegressor(n_estimators=n_est, max_depth=10, bootstrap=True, random_state=1)
    rf.fit(X_train, y_train)
    scores.append(rf.score(X_test, y_test))


plt.figure(figsize=(15, 10))
plt.xlabel("Maksymalna liczba lisci")
plt.ylabel("Wynik")

plt.plot(scores, 'm--')
#plt.plot(estimators 'b')

```

```python
Od n_estimators = 40 i w gore score praktycznie sie nie zmienia.
```

* Dopasuj model Random Forest z 80 estymatorami (parametr n_estimators) [20%]

Narysuj jak współczynnik R^2 zmienia się ze zmianą parametru `max_leaf_nodes` dla zbioru treningowego i testowego.

Użyj zakresu wartości: `range(2, 500, 10)`

```python
#from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size = 0.6 )
```

```python
rf_regressor = RandomForestRegressor(n_estimators = 80, max_depth=4, bootstrap=True, random_state=1)
rf_regressor.fit(X_train, y_train)

#rf_classifier = RandomForestClassifier(n_estimators = 80, max_depth=4, bootstrap=True, random_state=1)

```

```python
rf_train_scores, rf_test_scores = [], []


for i in range(2, 500, 10):
    rf_regressor = RandomForestRegressor(n_estimators = 80, max_leaf_nodes=i, max_depth=4, bootstrap=True, random_state=1)
    rf_regressor.fit(X_train, y_train)
    rf_regressor.fit(X_test, y_test)
    rf_train_scores.append(rf_regressor.score(X_train, y_train))
    rf_test_scores.append(rf_regressor.score(X_test, y_test))


plt.figure(figsize=(15, 10))
plt.xlabel("Maksymalna liczba lisci")
plt.ylabel("Wynik")

plt.plot(range(2, 500, 10), rf_train_scores, 'r')
plt.plot(range(2, 500, 10), rf_test_scores, 'b')

```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, accuracy_score, make_scorer
```

```python
boston = load_boston()
all_x, all_y = boston.data, boston.target
```

```python
param_grid = {"max_depth": [3, None],
              "max_features": list(range(1,10))
             }
print(param_grid)
```

```python
rf_classifier = RandomForestClassifier()

grid_search = GridSearchCV(rf_classifier, param_grid=param_grid, cv=5, verbose=0)
grid_search.fit(all_x, all_y)
```

```python
grid_search.cv_results_
```

```python
best_param_no = grid_search.best_index_ 
print('Best params set:', grid_search.best_params_, '(no. {0})'.format(best_param_no))
print('Best params score:', grid_search.best_score_ )
```
